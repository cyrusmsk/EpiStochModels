# SEIR model

```python
import epistochmodels
```

```python
pop = 1000
beta = 0.7
gamma = 0.2
e = 0.3
model = epistochmodels.SEIR(pop, beta, gamma, e)
```

```python
inits = 990,10,0,0
model.initialize(*inits)
```

```python
%%time
res = model.run(0,1000)
```

```python
%pylab inline
```

```python
plot(res[0],res[1])
legend(['S','E','I','R']);
```
